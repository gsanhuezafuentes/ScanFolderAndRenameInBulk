```python
import glob
import os
import re

import file_utils
```

```python
video_format = ".mp4 .avi .mkv".split(" ")
excluded_folders = {
    "D:\\Mis Cosas\\Descargas\\Nueva Carpeta",
    "D:\\Mis Cosas\\Descargas\\Nueva carpeta (2)",
    "D:\\Mis Cosas\\Descargas\\Peliculas",
    "D:\\Mis Cosas\\Descargas\\Final Fantasy XV"
}
descargas = "D:\Mis Cosas\Descargas"
```

```python
files = glob.glob(descargas+"/**", recursive=True)

filtered_files = file_utils.run_filter_pipeline(files, [
    lambda filepath: os.path.splitext(filepath)[1] in video_format,
    lambda filepath: file_utils.exclude_paths_in_exclude_list(filepath, excluded_folders),
    os.path.isfile
])

filtered_files[:15]
```

```python
# delete string using str.replace
modified_paths = file_utils.run_replacement(
    filtered_files,
    (
        (undesire_str, "")
        for undesire_str in (
            '(1080p Mkv)',
            '(1280x720 H264 AAC)',
            '(1280x720 x264  AAC)',
            '(1280x720 x264 10bits AAC)',
            '(1280x720 x264 AAC)',
            '(1280x720 x264)',
            '(1280x720)',
            '(1280x720+x264+AAC)',
            '(1280x720_H264_AAC)',
            '(1280x720_x264_AAC)',
            '(1280x720p H264 AAc)',
            '(1280x720p H264+AAc)',
            '(1280x720p_H264_AAc)',
            '(18)',
            '(2014)',
            '(2016)',
            '(2020)',
            #'(22)',
            #'(23)',
            #'(24)',
            #'(30-dai)',
            '(480p Mkv)',
            '(480p)',
            '(720)',
            '(720P)',
            '(720p HD)',
            '(720p Mkv BD-Rip)',
            '(720p Mkv)',
            '(720p)',
            '(720x480_h.264_aac)',
            '(848x480 x264 AAC)',
            '(AT-X 1280x720 x264 AAC)',
            '(BD 1280x720 x264 AAC)',
            '(BD 1920x1080 x264 AAC)',
            '(BDRip 1280x720p AC3)',
            '(BDrip 1920x1080 H264 AAC)',
            '(BDrip 1920x1080 x264 Hi444p)',
            '(BS11 1280x720 x264 AAC)',
            '(DVD 1280x720 x264 10bits AAC)',
            '(DVD 720x480 AVC AAC)',
            '(DVD 848x480 x264 AAC)',
            '(DVD-HD 1280x720 x264 AAC)',
            '(DVDrip 1280x720 H264 AAC)',
            '(Danmachi)',
            '(HD v2)',
            '(HD)',
            '(Japan-Paw.net)',
            '(Kuro)',
            '(MP4)',
            '(MX 1280x720 x264 AAC)',
            '(Mega-Anime)',
            '(NHK-E_1280x720_x264_AAC)',
            #'(OVA)',
            '(SD)',
            '(Spanish Sub)',
            '(Sub Esp)',
            #'(Thousand-Year Blood War)',
            #'(Zen)'
            ".720p-hi8",
            " By Julian12100",
            " By Julian 12100"
        )
    ),
)

# delete string using regex
modified_paths = file_utils.run_replacement(
    modified_paths,
    (
        (r"(\[.*?\])", ""),
        (fr'\s*(?=({"|".join([re.escape(extension) for extension in video_format])})$)', ""),  # elimina los espacios en blanco antes de la extensión r'\s(?=(\.mp4|\.avi|\.mkv)$)'
        #(r"(\(Sub Esp\)|\(MP4\)|\.720p\-hi8)", ""),
        (r"\s{2,}", " "),
    ),
    use_regex=True
)
modified_paths
```

```python
table_of_replacements = []
for original_path, modified_path in zip(filtered_files, modified_paths):
    filename = os.path.basename(original_path)
    modified_name, extension = os.path.basename(modified_path).rsplit(".", 1)
    table_of_replacements.append(
        {
            "full_path": original_path,
            "filename": filename,
            "new_name": modified_name.strip(),
            "extension": extension
        }
    )
table_of_replacements[:2]
```

```python
file_utils.save_in_csv(
    "replace.csv",
    ["full_path", "filename", "new_name", "extension"],
    table_of_replacements
)
```
